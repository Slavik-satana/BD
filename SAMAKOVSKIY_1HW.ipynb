# База данный в DBeaver

Ниже представлен код (SQL), с помщью которого создавались таблицы в DBeaver

```python
CREATE TABLE customers_address (
    customer_id SERIAL PRIMARY KEY,
    address TEXT,
    postcode INTEGER,
    state VARCHAR,
    country VARCHAR,
    property_valuation INTEGER
);

CREATE TABLE customers (
    customer_id SERIAL PRIMARY KEY,
    first_name TEXT,
    last_name TEXT,
    gender VARCHAR,
    DOB VARCHAR(10),
    job_title TEXT,
    job_industry_category VARCHAR,
    wealth_segment VARCHAR,
    deceased_indicator VARCHAR,
    owns_car VARCHAR,
    FOREIGN KEY (customer_id) REFERENCES customers (customer_id)
);

CREATE TABLE products (
    product_id SERIAL PRIMARY KEY,
    brand TEXT,
    product_line VARCHAR,
    product_class VARCHAR,
    product_size VARCHAR
);

CREATE TABLE transactions (
    transaction_id SERIAL PRIMARY KEY,
    product_id INTEGER,
    customer_id INTEGER,
    transaction_date DATE,
    online_order BOOLEAN,
    order_status VARCHAR(9),
    list_price FLOAT,
    standard_cost FLOAT,
    FOREIGN KEY (product_id) REFERENCES products (product_id),
    FOREIGN KEY (customer_id) REFERENCES customers (customer_id)
);
```

Поскольку одному product_id могут соответствовать несколько наборов значений, таблицы с транзакциями и продуктами записаны в обрезаном формате. Таблицы с покупателями и их адресами записаны полностью.
